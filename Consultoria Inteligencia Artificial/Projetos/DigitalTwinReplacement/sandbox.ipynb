```python
import json
import pandas as pd


json_file = 'SimulationResult.json'

with open(json_file) as f:
    data = json.load(f)

raw_dataframe = pd.DataFrame(data)
raw_dataframe.head()
```

Dados de Entrada do Modelo:
- Vazão do vapor gerado pelas caldeiras.
- Pressão do vapor gerado pelas caldeiras.
- Temperatura do vapor gerado pelas caldeiras.
- Vazão de vapor na entrada da turbina 1_1.
- Vazão de vapor na entrada da turbina 2_1.
- Status turbinas 1_1/1_2 (On/Off)
- Status turbinas 2_1/2_2 (On/Off)

Dados de Saída do Modelo:
- Consumo específico turbina 1_1.
- Consumo específico turbina 1_2.
- Consumo específico turbina 2_1.
- Consumo específico turbina 2_2.
- Potencia gerada turbina 1_1.
- Potencia gerada turbina 1_2.
- Potencia gerada turbina 2_1.
- Potencia gerada turbina 2_2.
- Vazão total de vapor de escape gerada.
- Pressão de vapor de escape gerada.
- Temperatura de vapor de escape gerada.

Dados Auxiliares do experimento (podem ser desconsiderados):
- Step corrente.
- Status da simulação.

```python
# Removendo dados com falha na simulação
non_nan_df = raw_dataframe.dropna()
non_nan_df.status.unique()
```

```python
non_nan_df.describe()
```

```python
# Valores negativos em potenciaGeradaTG2_2 são erros de simulação, vamos remover esses valores
non_nan_df.query('potenciaGeradaTG2_2 < 0').count()
```

```python
non_nan_df = non_nan_df[non_nan_df['potenciaGeradaTG2_2'] >= 0]
```

```python
# Como podemos ver temperaturaVaporEscape e pressaoVaporEscape são constantes, logo podemos removê-las
# Além disso, como a nossa metrica de avialação escolhida é o MAPE precisamos remover os valores iguais a zero das nossas variaveis alvo. Vamos fazer isso adicionando um offset de 10
```

```python
fature_columns = [
    'vazaoVapor', 'pressaoVapor', 'temperaturaVapor', 
    'cargaVaporTG1', 'cargaVaporTG2', 'habilitaTG1', 'habilitaTG2'
]
target_columns = [
    'potenciaGeradaTG1_1', 'potenciaGeradaTG1_2', 'potenciaGeradaTG2_1', 
    'potenciaGeradaTG2_2', 'vazaoVaporEscape', 'consumoEspecificoTG1_1', 'consumoEspecificoTG1_2', 
    'consumoEspecificoTG2_1', 'consumoEspecificoTG2_2'
]

dataframe = non_nan_df[fature_columns + target_columns]
dataframe.head()
```

```python
dataframe[target_columns] = dataframe[target_columns] + 10
dataframe.describe()
```

```python
# Separando dados em treino e teste
from sklearn.model_selection import train_test_split

features = dataframe[fature_columns]
targets = dataframe[target_columns]

features_train, features_test, targets_train, targets_test = train_test_split(
    features, targets, test_size=0.2, random_state=0
)

features_train.shape, features_test.shape, targets_train.shape, targets_test.shape
```

```python
import tensorflow as tf
from tensorflow.keras import Input, layers, models

class CustomModel:
    def __init__(self, input_layer_size: int, layer_sizes: list, output_size: int, loss='mean_squared_error', optimizer='adam', device: str = 'CPU:0'):
        """
        Initializes the model based on provided layer sizes.
        
        :param input_layer_size: The size of the input layer.
        :param layer_sizes: A list of integers where each integer specifies
                            the number of units in the layer.
        :param output_size: The size of the output layer.
        :param loss: Loss function to be used for model training.
        :param optimizer: Optimizer to be used for model training.
        """
        self.input_layer_size = input_layer_size
        self.layer_sizes = layer_sizes
        self.output_size = output_size
        self._loss = loss
        self._optimizer = optimizer
        self._metrics = ['mean_squared_error']  # Default metric for regression
        self._model = self._build_model()
        
    def set_metrics(self, metrics):
        self._metrics = metrics
        
    def _build_model(self):
        inputs = Input(shape=(self.input_layer_size,))
        x = inputs
        
        for size in self.layer_sizes:
            x = layers.Dense(size, activation='relu')(x)
        
        outputs = layers.Dense(self.output_size, activation='linear')(x)
        
        return models.Model(inputs=inputs, outputs=outputs)
        
    def compile_model(self):
        self._model.compile(
            optimizer=self._optimizer,
            loss=self._loss,
            metrics=self._metrics
        )
        
    def summary(self):
        return self._model.summary()
    
    @property
    def model(self):
        return self._model
    
    def fit(self, x, y, **kwargs):
        return self.model.fit(x, y, **kwargs)
    
    def evaluate(self, x, y):
        return self.model.evaluate(x, y)
```

```python
# Treinando o modelo

input_layer_size = features_train.shape[1]  # Number of features
layer_sizes = [64, 64]  # Example architecture: Two hidden layers with 64 neurons each
output_size = targets_train.shape[1]  # Number of targets to predict
model = CustomModel(input_layer_size, layer_sizes, output_size, loss='mean_squared_error', optimizer='adam')

model.set_metrics(['mae', 'mse'])  # Tracking both Mean Absolute Error and Mean Squared Error
model.compile_model()

model.summary()
```

```python
history = model.fit(features_train, targets_train, validation_split=0.2, epochs=100, batch_size=32)

test_loss, test_mae, test_mse = model.evaluate(features_test, targets_test)
print(f'Test Loss: {test_loss}, Test MAE: {test_mae}, Test MSE: {test_mse}')
```

```python
import numpy as np

predictions = model.model.predict(features_test)
errors = np.abs((targets_test - predictions) / targets_test) * 100

# Calculando MAPE ignorando NaNs
mape = np.nanmean(errors)
print(f'MAPE: {mape:.2f}%')
```
